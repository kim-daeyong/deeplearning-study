```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def diff(x, y, a, b, x_data):

    y_pred = a * x_data + b
    error = y_data - y_pred

    # a 편미분
    a_diff = -(2 / len(x_data) * sum(x_data * (error)))
    # b 편미분
    b_diff = -(2 / len(x_data) * sum(error))

    # 미분 결과에 학습률을 곱한 후 기존의 a, b 값을 업데이트
    a = a - lr * a_diff 
    b = b - lr * b_diff
```

```python
# 경사 하강법 (gradient descent)
data = [[2, 81], [4, 93], [6, 91], [8, 97]]
x = [i[0] for i in data]
y = [i[1] for i in data]

# 그래프
plt.figure(figsize=(8, 5))
plt.scatter(x, y)
plt.show()

# 리스트로 되어 있는 x와 y 값을 넘파이 배열로 바꾸기(인덱스를 주어 하나씩 불러와 계산이 가능하게 하기 위함)
x_data = np.array(x)
y_data = np.array(y)

# 기울기 a와 절편 b의 값 초기화
a = 0
b = 0

# 학습률 정하기
lr = 0.05

# 몇번 반복될지 설정
epochs = 2001

# 경사 하강법 시작
for i in range(epochs):
    y_pred = a * x_data + b # y 구하기
    error = y_data - y_pred # 오차 구하기
    
#     # 오차 함수를 a로 미분한 값
#     a_diff = -(2 / len(x_data)) * sum(x_data * error)
#     # 오차 함수를 b로 미분한 값
#     b_diff = -(2 / len(x_data)) * sum(error)

    a_diff = -(1 / len(x_data)) * sum(x_data * (error)) # 평균 제곱 오차를 a로 미분한 값
    b_diff = -(1 / len(x_data)) * sum(error) # 평균 제곱 오차를 b로 미분한 값
    
    a = a - lr * a_diff
    b = b - lr * b_diff
    
    if i % 100 == 0:
        print("epoch = %.f, 기울기 = %.04f, 절편 = %.04f" % (i, a, b))
        
y_pred = a * x_data + b
plt.scatter(x, y)
plt.plot([min(x_data), max(x_data)], [min(y_pred), max(y_pred)])
plt.show()
```
