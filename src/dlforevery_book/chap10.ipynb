```python
# 딥러닝을 구동하는 데 필요한 케라스 함수
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

import numpy as np
import tensorflow as tf

# 실행할 때마다 같은 결과를 출력하기 위해 설정.
np.random.seed(3)
tf.random.set_seed(3)

Data_set = np.loadtxt("../../../../../dlevery/dataset/ThoraricSurgery.csv", delimiter=",")

# 환자의 기록과 수술 결과를 X와 Y.
X = Data_set[:,0:17]
Y = Data_set[:,17]

# 두개층을 가진 모델, 은닉층과 마지막은 출력층
model = Sequential()
model.add(Dense(30, input_dim=17, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

# 딥러닝 실행
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(X, Y, epochs=100, batch_size=10) # loss: 0.3832 - accuracy: 0.8468
# # model.fit(X, Y, epochs=100, batch_size=20) # loss: 0.3909 - accuracy: 0.8532
# # model.fit(X, Y, epochs=100, batch_size=30) # loss: 0.3896 - accuracy: 0.8574
# # model.fit(X, Y, epochs=100, batch_size=40) # loss: 0.4052 - accuracy: 0.8532
# # model.fit(X, Y, epochs=100, batch_size=50) # loss: 0.4078 - accuracy: 0.8511

# model.compile(loss='mean_squared_error', optimizer='adam', metrics=['accuracy'])
# model.fit(X, Y, epochs=100, batch_size=10) # loss: 0.1297 - accuracy: 0.8553

### modelcheckpoint 로 loss 값이나 accuracy등으로 그 시점을 저장 할 수 있음!
```
